```python
def ramsey(num_idens, state):
    circ = qk.QuantumCircuit(2,1)
    theta, phi, lamb = state
    circ.u(theta, phi, lamb, 1)
    for _ in range(num_idens):
        for _ in range(51):
            circ.i(1)
            circ.i(1)
            circ.i(1)
            circ.i(1)
    circ.u(-theta, -lamb, -phi, 1)
    circ.measure(1,0)
    return circ
```

```python
def xy4(num_reps, state):
    circ = qk.QuantumCircuit(2,1)
    theta, phi, lamb = state
    circ.u(theta, phi, lamb, 1)
    for _ in range(num_reps):
        for _ in range(51):
            circ.x(1) 
            circ.rz(pi/2, 1)
            circ.x(1)
            circ.rz(-pi/2,1)
            circ.x(1)
            circ.rz(pi/2, 1)
            circ.x(1)
            circ.rz(-pi/2,1)
    circ.u(-theta, -lamb, -phi, 1)
    circ.measure(1,0)
    return circ
```

```python
def circuit_builder(exp, reps, state):
    if exp == 'free':
        circ = ramsey(reps,state)
    elif exp == 'XY4':
        circ = xy4(reps,state)
    return circ
```

```python
'''
Builds initial state dictionary composed of the 6 poles and 10 random states.
-- Initial states given in terms of theta, phi, lambda for U3 operator
'''

state_dict = {}
state_dict['+'] = [np.pi, 0, 0]
state_dict['-'] = [-np.pi/2, 0, 0]
state_dict['i'] = [-np.pi/2, -np.pi/2, np.pi/2]
state_dict['-i'] = [np.pi/2, -np.pi/2, np.pi/2]
state_dict['0'] = [0, 0, 0]
state_dict['1'] = [np.pi, 0, 0] 


# theta range: [0,pi]
# phi, lambda range: [0, 2pi]
for i in range(10):
    unnormalized = np.random.rand(3)
    theta = np.pi*unnormalized[0]
    phi = 2*np.pi*unnormalized[1]
    lamb = 2*np.pi*unnormalized[2]
    state_dict['rand%d' % i] = [theta, phi, lamb]

pk.dump(state_dict, open('random_states.p', 'wb'))
```

```python
batch_size = 96 # specify how many batches to run in a python console
batch_index = 0 # set index for notebook or console when running multiple
starting_circ_idx = batch_index*batch_size
final_circ_idx = starting_circ_idx + batch_size
num_batches = 5
```

```python
for curr_circ_idx in range(0, num_batches * batch_size, batch_size):


    batch_circuits = experiments[curr_circ_idx: curr_circ_idx + batch_size]
    circuits_to_run = {}

    for circ in batch_circuits:
        exp = circ['exp']
        reps = circ['reps']
        state_idx = circ['state']
        if state_idx in state_mapping:
            state_name = state_mapping[state_idx]
        state = state_dict[state_name]
        circuits_to_run['%s_reps%d_state%s' % (exp, reps, state_name)] = circuit_builder(exp, reps, state)
    print('Running circuits...')
    job = qk.execute(list(circuits_to_run.values()), backend=backend, shots=1000, optimization_level=0)
    
     # Extract essential information for storage
    job_id = job.job_id()
    circuit_specs = list(circuits_to_run.keys())

    circuit_meta_data = {
         'run date': datetime.now().strftime("%Y_%m_%d_%H%M%S"),
         'circuit specs': circuit_specs,
         'job_id': job_id
     }
    filename = '%s-ramseyVpdd-batch_%s_qubit1.p' % (backend, batch_index)
    print(filename)
    pk.dump(circuit_meta_data, open(filename, 'wb'))

    batch_index += 1
   
```

```python

# Initialize lists to store the average fidelity values for each state
free_state_avg = []
xy4_state_avg = []

# Iterate over the repetitions
for i in range(0,15):    
    # Initialize lists to store fidelity values for each state at a specific repetition
    free_state_values = []
    xy4_state_values = []

    # Iterate over the dictionary keys and extract fidelity values for each state
    for key, value in all_experiment_data.items():
        if key.startswith(f'free_reps{i}_state'):
            free_state_values.append(float(value['0']))
        elif key.startswith(f'XY4_reps{i}_state'):
            xy4_state_values.append(float(value['0']))
            
    # Calculate the average fidelity for each state at the current repetition
    free_state_avg.append(sum(free_state_values) / (len(free_state_values)*1000)) 
    xy4_state_avg.append(sum(xy4_state_values) / (len(xy4_state_values)*1000)) 

# Create x-axis values (repetitions)
repetitions = list(range(15))

# Create the plot
plt.plot(repetitions, free_state_avg, label='Free State')
plt.plot(repetitions, xy4_state_avg, label='XY4 State')

# Set axis labels and plot title
plt.xlabel('Repetitions')
plt.ylabel('Fidelity')
plt.title('Average Fidelity for Free State and XY4 State')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
